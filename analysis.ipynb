```python
mean_times = []
max_times = []
min_times = []
by_cities = df.groupby('#cities')
for key,item in by_cities:
    mean_times.append(item.time.mean())
    max_times.append(item.time.max())
    min_times.append(item.time.min())
    

time_by_cities = pd.DataFrame({'mean':mean_times,'max':max_times,'min':min_times}, index=set(df['#cities'].unique()) )

time_by_cities
```

```python
with open('df2.pickle', "rb") as input_file:
        df2 = pickle.load(input_file)
print(df2)

df2.frontier = df2.frontier.astype('int32')
df2.time = df2.time.astype('float')
#df.time.replace({-1.0:3600.0},inplace=True)
df2.generated = df2.generated.astype('int')
df2.expanded = df2.expanded.astype('int')
df2
```

```python
by_test = df2.groupby('#test')
gen_mean = []
exp_mean = []
time_mean = []
front_mean = []
it = 0

for key,item in by_test:
    gen_mean.append(item.generated.mean())
    exp_mean.append(item.expanded.mean())
    time_mean.append(item.time.mean())
    front_mean.append(item.frontier.mean())

    
info_by_test = pd.DataFrame({'Generated':gen_mean,'Expanded':exp_mean,'Frontier size':front_mean, 'Time(s)': time_mean}, index=df['#test'].unique())

info_by_test["Frontier size"] = info_by_test["Frontier size"].round(decimals = 0).astype('int')
info_by_test["Time(s)"] = info_by_test["Time(s)"].apply(lambda x: '{:.2f}'.format(x))
info_by_test['Generated'] = info_by_test['Generated'].apply(lambda x: '{:.0f}'.format(x))
info_by_test['Expanded'] = info_by_test['Expanded'].apply(lambda x: '{:.0f}'.format(x))
info_by_test
```

```python
print(info_by_test.to_latex(escape=False))
```
